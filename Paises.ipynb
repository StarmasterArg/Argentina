```
import pandas as pd

url = "https://raw.githubusercontent.com/StarmasterArg/DATASETS/main/countries%20of%20the%20world.csv"

df = pd.read_csv(url)

df.info()
df.head()
```

```
def clean_col(col):
    col = col.strip()
    col = col.replace(")","")
    col = col.replace("(","")
    col = col.replace(" ","_")
    col = col.lower()

    return col

df.columns = [clean_col(c) for c in df.columns]

df.columns
```

```
df["region"] = df["region"].apply(clean_col)
df["region"].unique()
```

```
ame = df[df['region'] == "latin_amer._&_carib"]

ame.info()
ame.head()
```

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
plt.figure(figsize=(18, 6))

plt.style.use("fivethirtyeight")
# plt.xkcd()
plt.plot(ame['country'],ame['gdp_$_per_capita'],color = "b", linestyle ="--",marker=".")

plt.title("Pbi en dolares por pais")
plt.xlabel("Paises")

plt.xticks(rotation=90)
plt.grid(False)


plt.show()
```
